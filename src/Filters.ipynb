```python
def crimmins(data):
    new_image = data.copy()
    nrow = len(data)
    ncol = len(data[0])
    
    # Dark pixel adjustment
    
    # First Step
    # N-S
    for i in range(1, nrow):
        for j in range(ncol):
            if data[i-1,j] >= (data[i,j] + 2):
                new_image[i,j] += 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(ncol-1):
            if data[i,j+1] >= (data[i,j] + 2):
                new_image[i,j] += 1
    data = new_image
    # NW-SE
    for i in range(1, nrow):
        for j in range(1, ncol):
            if data[i-1,j-1] >= (data[i,j] + 2):
                new_image[i,j] += 1
    data = new_image
    #NE-SW
    for i in range(1, nrow):
        for j in range(ncol-1):
            if data[i-1,j+1] >= (data[i,j] + 2):
                new_image[i,j] += 1
    data = new_image
    # Second Step
    # N-S
    for i in range(1, nrow-1):
        for j in range(ncol):
            if (data[i-1,j] > data[i,j]) and (data[i,j] <= data[i+1,j]):
                new_image[i,j] += 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(1, ncol-1):
            if (data[i,j+1] > data[i,j]) and (data[i,j] <= data[i,j-1]):
                new_image[i,j] += 1
    data = new_image
    # NW-SE
    for i in range(1, nrow-1):
        for j in range(1, ncol-1):
            if (data[i-1,j-1] > data[i,j]) and (data[i,j] <= data[i+1,j+1]):
                new_image[i,j] += 1
    data = new_image
    # NE-SW
    for i in range(1, nrow-1):
        for j in range(1, ncol-1):
            if (data[i-1,j+1] > data[i,j]) and (data[i,j] <= data[i+1,j-1]):
                new_image[i,j] += 1
    data = new_image
    #Third Step
    # N-S
    for i in range(1, nrow-1):
        for j in range(ncol):
            if (data[i+1,j] > data[i,j]) and (data[i,j] <= data[i-1,j]):
                new_image[i,j] += 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(1, ncol-1):
            if (data[i,j-1] > data[i,j]) and (data[i,j] <= data[i,j+1]):
                new_image[i,j] += 1
    data = new_image
    # NW-SE
    for i in range(1, nrow-1):
        for j in range(1, ncol-1):
            if (data[i+1,j+1] > data[i,j]) and (data[i,j] <= data[i-1,j-1]):
                new_image[i,j] += 1
    data = new_image
    # NE-SW
    for i in range(1, nrow-1):
        for j in range(1, ncol-1):
            if (data[i+1,j-1] > data[i,j]) and (data[i,j] <= data[i-1,j+1]):
                new_image[i,j] += 1
    data = new_image
    # Fourth Step
    # N-S
    for i in range(nrow-1):
        for j in range(ncol):
            if (data[i+1,j] >= (data[i,j]+2)):
                new_image[i,j] += 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(1,ncol):
            if (data[i,j-1] >= (data[i,j]+2)):
                new_image[i,j] += 1
    data = new_image
    # NW-SE
    for i in range(nrow-1):
        for j in range(ncol-1):
            if (data[i+1,j+1] >= (data[i,j]+2)):
                new_image[i,j] += 1
    data = new_image
    # NE-SW
    for i in range(nrow-1):
        for j in range(1,ncol):
            if (data[i+1,j-1] >= (data[i,j]+2)):
                new_image[i,j] += 1
    data = new_image
    
    # Light pixel adjustment
    
    # First Step
    # N-S
    for i in range(1,nrow):
        for j in range(ncol):
            if (data[i-1,j] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(ncol-1):
            if (data[i,j+1] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    # NW-SE
    for i in range(1,nrow):
        for j in range(1,ncol):
            if (data[i-1,j-1] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    # NE-SW
    for i in range(1,nrow):
        for j in range(ncol-1):
            if (data[i-1,j+1] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    # Second Step
    # N-S
    for i in range(1,nrow-1):
        for j in range(ncol):
            if (data[i-1,j] < data[i,j]) and (data[i,j] >= data[i+1,j]):
                new_image[i,j] -= 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(1, ncol-1):
            if (data[i,j+1] < data[i,j]) and (data[i,j] >= data[i,j-1]):
                new_image[i,j] -= 1
    data = new_image
    # NW-SE
    for i in range(1,nrow-1):
        for j in range(1,ncol-1):
            if (data[i-1,j-1] < data[i,j]) and (data[i,j] >= data[i+1,j+1]):
                new_image[i,j] -= 1
    data = new_image
    # NE-SW
    for i in range(1,nrow-1):
        for j in range(1,ncol-1):
            if (data[i-1,j+1] < data[i,j]) and (data[i,j] >= data[i+1,j-1]):
                new_image[i,j] -= 1
    data = new_image
    # Third Step
    # N-S
    for i in range(1,nrow-1):
        for j in range(ncol):
            if (data[i+1,j] < data[i,j]) and (data[i,j] >= data[i-1,j]):
                new_image[i,j] -= 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(1,ncol-1):
            if (data[i,j-1] < data[i,j]) and (data[i,j] >= data[i,j+1]):
                new_image[i,j] -= 1
    data = new_image
    # NW-SE
    for i in range(1,nrow-1):
        for j in range(1,ncol-1):
            if (data[i+1,j+1] < data[i,j]) and (data[i,j] >= data[i-1,j-1]):
                new_image[i,j] -= 1
    data = new_image
    # NE-SW
    for i in range(1,nrow-1):
        for j in range(1,ncol-1):
            if (data[i+1,j-1] < data[i,j]) and (data[i,j] >= data[i-1,j+1]):
                new_image[i,j] -= 1
    data = new_image
    # Fourth Step
    # N-S
    for i in range(nrow-1):
        for j in range(ncol):
            if (data[i+1,j] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    # E-W
    for i in range(nrow):
        for j in range(1,ncol):
            if (data[i,j-1] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    # NW-SE
    for i in range(nrow-1):
        for j in range(ncol-1):
            if (data[i+1,j+1] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    # NE-SW
    for i in range(nrow-1):
        for j in range(1,ncol):
            if (data[i+1,j-1] <= (data[i,j]-2)):
                new_image[i,j] -= 1
    data = new_image
    return new_image.copy()
```

**List of different noise removal filters**
Each filter can take image and will apply the filter techniques and it will save the filtered image in each seperate folder 

```python
class ListOfFilters:
    images=[] # list contatining  all images
    def applyMedianFilter(self,img,filename):
        #print("Median Filter")
        figure_size = 9 # the dimension of the x and y axis of the kernal.
        #img = np.full((512,512,3), 12, np.uint8)
        image = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)
        image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        median_img = cv2.medianBlur(image2, figure_size)
        # Saves image of the current frame in jpg file
        name = "./Median/Median_"+filename
        #print ('Creating...' + name)
        cv2.imwrite(name, median_img)

    def applyMeanFilter(self,img,filename):
        #print("Mean Filter")
        image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) # convert to HSV
        figure_size = 9 # the dimension of the x and y axis of the kernal.
        mean_img = cv2.blur(image,(figure_size, figure_size))
        # Saves image of the current frame in jpg file
        name = './Mean/Mean_'+filename
        #print ('Creating...' + name)
        cv2.imwrite(name, mean_img)

    def conservative_smoothing_gray(self, data, filter_size):
        
        temp = []

        indexer = filter_size // 2

        new_image = data.copy()

        nrow, ncol = data.shape

        for i in range(nrow):

            for j in range(ncol):

                for k in range(i-indexer, i+indexer+1):

                    for m in range(j-indexer, j+indexer+1):

                        if (k > -1) and (k < nrow):

                            if (m > -1) and (m < ncol):

                                temp.append(data[k,m])

                temp.remove(data[i,j])


                max_value = max(temp)

                min_value = min(temp)

                if data[i,j] > max_value:

                    new_image[i,j] = max_value

                elif data[i,j] < min_value:

                    new_image[i,j] = min_value

                temp =[]

        return new_image.copy()

    def applyConservativeFilter(self,img,filename):
        #print("Conservative Filter")
        image = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)
        image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        conv_img = self.conservative_smoothing_gray(image2,5)
        # Saves image of the current frame in jpg file
        name = './Conservative/Conservative_'+filename
        #print ('Creating...' + name)
        cv2.imwrite(name, conv_img)

    def applyUnsharpFilter(self,img,filename):
        #print("Unsharp Filter")
        image = Image.fromarray(img.astype('uint8'))
        unsharp_img = image.filter(ImageFilter.UnsharpMask(radius=2, percent=150))
         # Saves image of the current frame in jpg file
        name = './Unsharp/Unsharp_'+filename
        #print ('Creating...' + name)
        cv2.imwrite(name, np.array(unsharp_img))

    def applySpeckleFilter(self,img,filename):
        #print("Speckle Filter")
        image = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)
        image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        new_image = crimmins(image2)
         # Saves image of the current frame in jpg file
        name = './Speckle/Speckle_'+filename
        #print ('Creating...' + name)
        cv2.imwrite(name, new_image)
        
    def printImage(self, image2, filtered_image):
        plt.figure(figsize=(11,6))
        plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
        plt.xticks([]), plt.yticks([])
        plt.subplot(122), plt.imshow(filtered_image, cmap='gray'),plt.title('Filtered Image')
        plt.xticks([]), plt.yticks([])
        plt.show()
```

**Calling Filters by giving image folder**

```python

class Filters(ListOfFilters):
    def __init__(self,folderName):
        #image folder name 
        #folder="E:/LHU/text_books/Project/research_proposal/code/output/MeanFilter/"
        self.folderName=folderName
        
    def applyImageFilters(self):
        
            if not os.path.exists('Median'):
                os.makedirs('Median')
            if not os.path.exists('Mean'):
                os.makedirs('Mean')
            if not os.path.exists('Unsharp'):
                os.makedirs('Unsharp')
            if not os.path.exists('Speckle'):
                os.makedirs('Speckle')
            if not os.path.exists('Conservative'):
                os.makedirs('Conservative')

            for filename in tqdm(os.listdir(self.folderName)):
                #print(self.folderName+"/"+filename)
                img=plt.imread(self.folderName+"/"+filename)  # reading image (Folder path and image name )
                super(Filters, self).applyMedianFilter(img,filename)
                super(Filters, self).applyMeanFilter(img,filename)
                super(Filters, self).applySpeckleFilter(img,filename)
                super(Filters, self).applyConservativeFilter(img,filename)
                super(Filters, self).applyUnsharpFilter(img,filename)
           
```
